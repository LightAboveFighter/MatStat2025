```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = [-1.11, -6.10, 2.42]
y = [-2.29, -2.91]

alpha = 0.05
N = 5000
beta = 1 - alpha
h_wave = np.mean(x) - np.mean(y)
h = 0
delta = h_wave - h
```

```python
bst_delta = []
for i in range(N):
    bst_delta.append(np.mean(np.random.choice(x, size=len(x), replace=True)) - np.mean(np.random.choice(y, size=len(y), replace=True)) - h_wave)

d_i = sorted(bst_delta)
print("Доверительный интервал: h <", round(h_wave - d_i[int(N * (1 - beta))], 3))

t = np.linspace(0, 10, 100)
F = [1 - norm.cdf(h_wave - d_i[int(N * (1 - beta))], loc=i, scale=np.sqrt(3)) for i in t]
plt.figure(figsize=(8, 6))
plt.plot(t, F)
plt.ylabel('W')
plt.xlabel('Δ = μ₁ - μ₂')
plt.grid()
plt.show()
```
